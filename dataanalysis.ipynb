```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df['Make_Encode']=lb.fit_transform(df['Make'])
```

```python
df
```

```python
feature = {}
for k in df.columns:
    if df[k].dtypes=='O':
        df[str(k)+'_Encode'] = lb.fit_transform(df[k])
        gt = {}
        ke = df[[k,str(k)+'_Encode']].drop_duplicates()[str(k)+'_Encode'].values
        val = df[[k,str(k)+'_Encode']].drop_duplicates()[k].values
        for i in range(len(ke)):
            gt[ke[i]] = val[i]
        feature[k] = gt
```
